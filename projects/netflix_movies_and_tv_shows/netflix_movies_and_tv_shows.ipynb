## Data Cleaning

```python
df['date_added']
```

```python
def month_extract(date_str):
  month= date_str.strip().split(' ')[0]
  month = month.lower()
  month_dic = {'Unknown':-1, 'january':1, 'february':2, 'march':3, 'april':4, 'may':5, 'june':6, 'july':7, 'august':8, 'september':9, 'october':10, 'november':11, 'december':12}
  return month_dic[month]
```

```python
df['date_added'][df['date_added'].notna()]
```

```python
ds_date = df['date_added'][df['date_added'].notna()]
ds_date = pd.to_datetime({'year': ds_date.apply(lambda date_str: int(date_str.strip().split(' ')[2])),
                          'day': ds_date.apply(lambda date_str: int(date_str.strip().split(' ')[1][:-1])),
                          'month': ds_date.apply(month_extract)},
                         format='%Y-%m-%d')

df.loc[df['date_added'].notna(), 'date_added'] = ds_date.dt.date
```

## What are the most common genres listed?

```python
df.head()
```

### data in normalized form

```python
df_temp = pd.DataFrame([], columns= df.columns)
```

```python
df_temp
```

```python
for row in df.iterrows():
  for genre in row[1].iloc[9].split(','):
    df_temp.loc[len(df_temp)] = row[1]
    df_temp.loc[len(df_temp)-1, 'listed_in'] = genre

df_temp.head(10)
```

```python
df_temp2 = df.copy()
```

```python
# Step 1: Split the 'listed_in' column into lists
df_temp2['listed_in'] = df_temp2['listed_in'].str.split(',')

# Step 2: Remove any leading/trailing whitespace from each genre
df_temp2['listed_in'] = df_temp2['listed_in'].apply(lambda genres: [genre.strip() for genre in genres])

# Step 3: Use explode to create a new row for each genre
df_temp2 = df_temp2.explode('listed_in', ignore_index=True)
```

```python
df_temp2.head(10)
```
